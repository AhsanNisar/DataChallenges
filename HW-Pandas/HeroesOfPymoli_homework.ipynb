```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data_df = pd.read_csv(file_to_load)

```

```python
# Calculation for the total players
total_players = len(purchase_data_df['SN'].unique())
# New data frame dispaying total players
total_players_df = pd.DataFrame({"Total Players": [total_players]}) 
total_players_df
```

```python
#Calculation of columns for Purchasing Analysis (Total)
unique_items = purchase_data_df['Item ID'].nunique()
average_price = purchase_data_df['Price'].mean()
purchase_num = len(purchase_data_df['Purchase ID'].unique())
total_revenue = purchase_data_df['Price'].sum()

#Summary table to display all the calculated values
summary_df = pd.DataFrame({"Number of Unique Items": [unique_items] ,
                          "Average Price": [average_price]  ,
                          "Number of Purchases": [purchase_num] ,
                          "Total Revenue": [total_revenue] }) 
#Formatting the newly created summary table 
summary_df["Average Price"] = summary_df["Average Price"].map("${:.2f}".format)
summary_df["Total Revenue"] = summary_df["Total Revenue"].map("${:,.2f}".format)
summary_df
```

```python
#Creating new data table with unique values for all the  'SN'
unique_data_df = purchase_data_df.drop_duplicates(subset='SN')

#Calculation of columns for Gender demographics
male = len(unique_data_df.loc[unique_data_df["Gender"] == "Male"])
female = len(unique_data_df.loc[unique_data_df["Gender"] == "Female"])
other = len(unique_data_df.loc[unique_data_df["Gender"] == "Other / Non-Disclosed"])

#Calculation of columns for Gender demographics percentage
male_percent = (len(unique_data_df.loc[unique_data_df["Gender"] == "Male"])/len(unique_data_df.loc[(unique_data_df["Gender"] == "Male") | (unique_data_df["Gender"] == "Female") |(unique_data_df["Gender"] == "Other / Non-Disclosed")]))*100
female_percent = (len(unique_data_df.loc[unique_data_df["Gender"] == "Female"])/len(unique_data_df.loc[(unique_data_df["Gender"] == "Male") | (unique_data_df["Gender"] == "Female") |(unique_data_df["Gender"] == "Other / Non-Disclosed")]))*100
other_percent = (len(unique_data_df.loc[unique_data_df["Gender"] == "Other / Non-Disclosed"])/len(unique_data_df.loc[(unique_data_df["Gender"] == "Male") | (unique_data_df["Gender"] == "Female") |(unique_data_df["Gender"] == "Other / Non-Disclosed")]))*100

#Creating new summary data frame to diaplay the calculated values
rows=["Male", "Female", "Other / Non-Disclosed"]
demographics_df = pd.DataFrame({"Total Count": [male, female, other] ,
                          "Percentage of Players": [male_percent, female_percent, other_percent]
                               },index=[rows])
#Formatting the newly created gender dempgraphics gender
demographics_df["Percentage of Players"] = demographics_df["Percentage of Players"].map("{:.2f}".format)
demographics_df
```

```python
# Calculation of purchaing analysis by Gender
purchase_value = purchase_data_df.groupby(["Gender"]).sum()
purchase_count = purchase_data_df.groupby(["Gender"]).count()
avg_purchase_price = purchase_data_df.groupby(["Gender"]).mean()
avg_total_per_person = purchase_data_df.groupby(["Gender"]).nunique()
avg_total_per_person = purchase_value['Price']/avg_total_per_person['SN']
purchase_value = purchase_value['Price']
purchase_count = purchase_count['Price']
avg_purchase_price = avg_purchase_price['Price']
    
#Creating a summary table to display the new values
purchasing_analysis = pd.DataFrame({"Purchase Count": purchase_count ,
                          "Average Purchase Price": avg_purchase_price  ,
                          "Total Purchase Value": purchase_value ,
                          "Avg Total Purchase per Person": avg_total_per_person }) 
#Formatting the newly created summary table 
purchasing_analysis["Average Purchase Price"] = purchasing_analysis["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis["Total Purchase Value"] = purchasing_analysis["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis["Avg Total Purchase per Person"] = purchasing_analysis["Avg Total Purchase per Person"].map("${:.2f}".format)



purchasing_analysis
```

```python
# creating the bins 
bins = [0,9,14,19,24,29,34,39,45]
group_names =["<10","10-14","15-19","20-24","25-29","30-34","35-39","40+"]

#putting the data in the bins 
purchase_data_df["Age Category"] = pd.cut(purchase_data_df["Age"], bins, labels=group_names)
#Calcukation of the values
agegroup = purchase_data_df.groupby(["Age Category"]).nunique()
agegroup_count = agegroup['SN']
agegroup_sum = agegroup_count.sum()
percent_players = (agegroup_count/agegroup_sum)*100

# Creattion of summary table to display age demographics
age_demographics_df = pd.DataFrame({"Total Count": agegroup_count ,
                          "Percentage of Players": percent_players
                                   }) 
#Formatting the data frame
age_demographics_df["Percentage of Players"] = age_demographics_df["Percentage of Players"].map("{:.2f}".format)
age_demographics_df
```

```python
#Calculation of new variables 
purchase_value_age = purchase_data_df.groupby(["Age Category"]).sum()
purchase_count_age = purchase_data_df.groupby(["Age Category"]).count()
avg_purchase_price_age = purchase_data_df.groupby(["Age Category"]).mean()
avg_total_per_person_age = purchase_data_df.groupby(["Age Category"]).nunique()
avg_total_per_person_age = purchase_value_age['Price']/avg_total_per_person_age['SN']
purchase_value_age = purchase_value_age['Price']
purchase_count_age = purchase_count_age['Price']
avg_purchase_price_age = avg_purchase_price_age['Price']
    
#Creating new sumamry data frame 
purchasing_analysis_age = pd.DataFrame({"Purchase Count": purchase_count_age ,
                          "Average Purchase Price": avg_purchase_price_age  ,
                          "Total Purchase Value": purchase_value_age ,
                          "Avg Total Purchase per Person": avg_total_per_person_age }) 

#Formatting the summary table 
purchasing_analysis_age["Average Purchase Price"] = purchasing_analysis_age["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_age["Total Purchase Value"] = purchasing_analysis_age["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_age["Avg Total Purchase per Person"] = purchasing_analysis_age["Avg Total Purchase per Person"].map("${:.2f}".format)



purchasing_analysis_age
```

```python
#Calculation of variables to calculate top spenders
purchase_value_sn = purchase_data_df.groupby(["SN"]).sum()
purchase_count_sn = purchase_data_df.groupby(["SN"]).count()
avg_purchase_price_sn = purchase_data_df.groupby(["SN"]).mean()
purchase_value_sn = purchase_value_sn['Price']
purchase_count_sn = purchase_count_sn['Price']
avg_purchase_price_sn = avg_purchase_price_sn['Price']
    
#Creating a summary data frame to input top spenders
purchasing_analysis_sn = pd.DataFrame({"Purchase Count": purchase_count_sn ,
                          "Average Purchase Price": avg_purchase_price_sn  ,
                          "Total Purchase Value": purchase_value_sn 
                                       }) 
#Sort by top purchase value column
purchasing_analysis_sn = purchasing_analysis_sn.sort_values(["Total Purchase Value"], ascending=False)

#Formatting the data frame 
purchasing_analysis_sn["Average Purchase Price"] = purchasing_analysis_sn["Average Purchase Price"].map("${:.2f}".format)
purchasing_analysis_sn["Total Purchase Value"] = purchasing_analysis_sn["Total Purchase Value"].map("${:,.2f}".format)

purchasing_analysis_sn.head()
```

```python
#Extracting the "Item ID", "Item Name", "Price" columns from the data frame and assining to a new data frame 
item_subset_df = purchase_data_df.filter([ "Item ID", "Item Name", "Price" ])
#Calculation of most popular items
purchase_count_subset = item_subset_df.groupby(["Item ID","Item Name"]).count() 
item_price_subset = item_subset_df.groupby(["Item ID","Item Name"]).mean() 
purchase_value_subset = item_subset_df.groupby(["Item ID","Item Name"]).sum()

purchase_count_subset = purchase_count_subset['Price']
item_price_subset = item_price_subset['Price']
purchase_value_subset = purchase_value_subset['Price']

#Creating a new data frame to dispay the most popular items 
purchasing_analysis_subset = pd.DataFrame({"Purchase Count": purchase_count_subset ,
                          "Item Price": item_price_subset  ,
                          "Total Purchase Value": purchase_value_subset 
                                       }) 
#Sorting  the Purchase count
purchasing_analysis_sorted = purchasing_analysis_subset.sort_values(["Purchase Count"], ascending=False)

#Formatting the values in the columns
purchasing_analysis_sorted["Item Price"] = purchasing_analysis_sorted["Item Price"].map("${:.2f}".format)
purchasing_analysis_sorted["Total Purchase Value"] = purchasing_analysis_sorted["Total Purchase Value"].map("${:,.2f}".format)
purchasing_analysis_sorted.head()
```
